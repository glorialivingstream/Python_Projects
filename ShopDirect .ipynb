This project involves a system called ShopDirect, which redirects a user to a specific online shop based on user's budget for clothing. The online shops used are as follows:
1. Amazon
2. Shein
3. Littleton
4. Boohoo
5. Pretty Little Thing

```python
# Creating a class for the online shops 
class ShopDirect:
    
    category = "clothing"
    
    def __init__(self, subcategory, quantity, budget):
        self.subcategory= subcategory
        self.quantity= quantity
        self.budget = budget
        
       
    def amazon(self):
        return f" Please visit Amazon using the link: www.amazon.com"
    
    def shein(self):
        return f" Please visit Shein using the link: www.shein.com"
    
    def littleton(self):
        return f" Please visit Littleton using the link: www.littleton.com"
    
    def boohoo(self):
        return f" Please visit Boohoo using the link: www.boohoo.com"
    
    def plt(self):
        return f" Please visit Pretty Little Thing using the link: www.plt.com"
    
        
```

 Using if statements to assess user's budget with subcategory to advise on the best shop choice.  
 The average price($) per clothing item in each shop is as follows:

 Littleton = (shirt: 5, shoe: 25, trousers: 10, hats: 2, other accessories : 5)
